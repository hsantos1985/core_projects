###Este notebook contem parte do estudo do meu TCC para o MBA
O objetivo dele é:

Criar um modelo de classificação que preve se o candidato da prova do ENEM, consegue tirar uma nota maior que a nota mínima de corte das principais universidades do seu Estado.

São mais de 30 variáveis explicativas para prever a variável target.
As bases utilizadas são referentes as provas do ENEM de 2021, para teste e treino do modelo e a prova de 2022, para validação.

De forma inicial, foram aplicados 9 modelos de classificação, sendo 6 de machine learning e 3 de redes neurais, que são:

--Regressão logística

--Árvore de decisão

--Random Forest

--XGBoost

--Catboost

--LGBM

Para redes neurais foram utilizados:

--Redes densas

--Redes Convolucionais

--Redes LSTM

O melhor modelo escolhido foi o LightGBM com acurácia de 99,93%.
As 15 variáveis mais importantes encontradas pelo modelo foram:

1--Nota final da prova

2--Média menor nota de corte das principais universidades públicas do Estado

3-Custo de mão de obra

4-Nota da prova de ciências humanas

5-Nota da prova de ciências naturais

6-Nota da prova de matemática

7-Formalidade do mercado de trabalho

8-Nota da prova de linguagens e códigos

9-Taxa de atendimento do ensino infantil

10-Proporção da PEA com Ensino superior

11-Inserção econômica

12-Produtividade do trabalho

13-Nota da redação

14-Taxa de frequência líquida no ensino fundamental

15-Segurança patrimonial



Considera-se que o modelo consegue de forma bem satisfatória prever se um determinado candidato irá obter a nota de corte mínima das principais universidades do seu Estado com uma acurácia de 99,93%

A nota final do aluno, assim como nas provas são determinantes neste modelo.
Todavia, o modelo também considera como essenciais parar aprovação dos candidatos , Estados que possuam altos rendimentos das pessoas ocupadas (Custo de mão de obra) e mercados de trabalho altamente formalizados (Formalização do mercado de trabalho)

##4.1-VERIFICAÇÃO DO BALANCEAMENTO DA BASE

```python
#verificação se a base esta balanceada
base_abt['TARGET'].value_counts()/len(base_abt)
#a base está desbalanceada, existem 70% de valores na classe 1 e 29% na classe 0
```

```python
#CRIANDO O BALANCEAMENTO DA BASE - UTILIZANDO O METODO UPSAMPLE
from sklearn.utils import resample

#separando as classes majoritarias
df_max=base_abt[base_abt['TARGET']==1]
```

```python
df_max
```

```python
#separando as classes minoritaria
df_min=base_abt[base_abt['TARGET']==0]
```

```python
df_min
```

```python
#utilizando o upsample pois aumneta o numero de casos da classe minoritaria
df_min_upsample=resample(
    df_min,
    replace=True,
    n_samples=len(df_max),
    random_state=42)
```

```python
#uniao das 2 classes
base_abt_up=pd.concat([df_max,df_min_upsample])
base_abt_up
```

```python
#embaralhando o df
base_abt_up=base_abt_up.sample(frac=1,random_state=42).reset_index(drop=True)
```

```python
#verificando se os dados estão balanceados
base_abt_up['TARGET'].value_counts()/len(base_abt_up)
```

```python
#SEPARAÇÃO DE TREINO EM 70% E TESTE 30%
from sklearn.model_selection import train_test_split
y=base_abt_up['TARGET']
X=base_abt_up[['TP_FAIXA_ETARIA', 'TP_SEXO', 'TP_ESTADO_CIVIL',
       'TP_COR_RACA', 'TP_NACIONALIDADE', 'TP_ST_CONCLUSAO', 'TP_ANO_CONCLUIU',
       'TP_ESCOLA', 'IN_TREINEIRO','NU_NOTA_CN',
       'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'TP_STATUS_REDACAO',
       'NU_NOTA_REDACAO', 'CMO', 'PEA_ES', 'PT', 'QT',
       'FMT', 'IE', 'IEJ', 'TFEF', 'TFEM', 'TAEI', 'PSPP', 'EGRPE', 'AST',
       'QST', 'CC', 'CEE', 'QEE', 'QR', 'BMD', 'TM', 'TC', 'CPFT', 'TI',
       'TMVI', 'SPAT', 'FALP', 'DR', 'DI', 'OI', 'CV',
       'MEDIA_MENOR_NOTA_CORTE', 'MEDIA_MAIOR_NOTA', 'regiao',
       'NOTA_FINAL']]


X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#aplicação base de validação
import pandas as pd
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score


df_val = pd.DataFrame(columns=['acc', 'precision', 'recall', 'f1', 'roc_auc'])

for type_algorithms, algorithms in algoritmos.items():
    # Definir as etapas do pipeline
    if type_algorithms == 'linear':
        steps = steps_comuns + steps_linear
    elif type_algorithms == 'tree':
        steps = steps_comuns

    for alg in algorithms:
        nome = alg[0]
        print(f'Validando o algoritmo {nome}', end=' ... ')

        # Criar o pipeline do modelo
        model_pipe = Pipeline(steps=steps + [alg])

        # Treinar o modelo
        model_pipe.fit(X_train, y_train)

        # Prever na base de validação
        y_pred = model_pipe.predict(X_val)

        # Para calcular o roc_auc é necessário ter as probabilidades
        y_proba = model_pipe.predict_proba(X_val)[:, 1]  # Proba da classe positiva

        # Calcular as métricas
        acc = accuracy_score(y_val, y_pred)
        precision = precision_score(y_val, y_pred, average='macro')
        recall = recall_score(y_val, y_pred, average='macro')
        f1 = f1_score(y_val, y_pred, average='macro')
        roc_auc = roc_auc_score(y_val, y_proba, multi_class='ovr')  # Supondo classificação binária

        # Adicionar os resultados ao DataFrame
        df_val.loc[nome] = [acc, precision, recall, f1, roc_auc]
        print('OK!')
```
